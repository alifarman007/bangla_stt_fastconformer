## Cell 2: Load the Hishab FastConformer Model

```
import nemo.collections.asr as nemo_asr
import torch
import warnings
warnings.filterwarnings('ignore')

# Check GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"üñ•Ô∏è Using device: {device}")
if device == "cuda":
    print(f"üéÆ GPU: {torch.cuda.get_device_name(0)}")
    print(f"üíæ VRAM: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

# Load model
print("\n‚è≥ Loading Hishab FastConformer model...")
asr_model = nemo_asr.models.ASRModel.from_pretrained("hishab/titu_stt_bn_fastconformer")
#If You want large model then comment out this
#asr_model = nemo_asr.models.ASRModel.from_pretrained("hishab/titu_stt_bn_conformer_large")
asr_model.eval()

print("\n‚úÖ Model loaded successfully!")
print("üìä Model size: ~463 MB")
print("üéØ Ready for Bangla speech recognition!")
```

## Cell 3: Helper Functions

```
import os
import tempfile
from pydub import AudioSegment
import numpy as np

def convert_to_wav_16k(input_path, output_path=None):
    """
    Convert any audio file to 16kHz mono WAV (required by model)
    """
    if output_path is None:
        output_path = tempfile.mktemp(suffix=".wav")

    # Load audio
    audio = AudioSegment.from_file(input_path)

    # Convert to mono and 16kHz
    audio = audio.set_channels(1)
    audio = audio.set_frame_rate(16000)

    # Export
    audio.export(output_path, format="wav")

    return output_path

def transcribe_audio(audio_path):
    """
    Transcribe audio file to Bangla text
    """
    # Convert to proper format
    wav_path = convert_to_wav_16k(audio_path)

    # Get duration
    audio = AudioSegment.from_wav(wav_path)
    duration = len(audio) / 1000  # seconds

    # Transcribe
    with torch.no_grad():
        transcriptions = asr_model.transcribe([wav_path])

    # Extract text from result
    if hasattr(transcriptions[0], 'text'):
        text = transcriptions[0].text
    else:
        text = str(transcriptions[0])

    # Cleanup temp file
    if wav_path != audio_path:
        os.remove(wav_path)

    return text, duration

print("‚úÖ Helper functions loaded!")
```

## üé§ Option B: Record from Microphone

**Instructions:**
1. Run the cell below
2. Click **"üé§ Start Recording"** button
3. Allow microphone access when prompted
4. Speak in Bangla
5. Click **"‚èπÔ∏è Stop Recording"** button
6. Wait for transcription

```
# ============================================
# üé§ Option B: Record from Microphone (FIXED)
# ============================================

from IPython.display import display, Javascript, HTML, Audio
from google.colab import output
from base64 import b64decode
from pydub import AudioSegment
import os

# Callback function to process audio
def process_audio(base64_audio):
    print("\n‚è≥ Processing recording...")

    # Decode base64
    audio_data = b64decode(base64_audio)

    # Save as webm
    webm_path = "/content/recorded_audio.webm"
    with open(webm_path, "wb") as f:
        f.write(audio_data)

    # Convert to WAV using pydub
    wav_path = "/content/recorded_audio.wav"
    audio = AudioSegment.from_file(webm_path, format="webm")
    audio = audio.set_channels(1).set_frame_rate(16000)
    audio.export(wav_path, format="wav")

    # Transcribe
    print("‚è≥ Transcribing your recording...")
    text, duration = transcribe_audio(wav_path)

    # Display results
    print("\n" + "=" * 50)
    print("üìù TRANSCRIPTION RESULT")
    print("=" * 50)
    print(f"\nüéØ Text: {text}")
    print(f"\n‚è±Ô∏è Recording duration: {duration:.2f} seconds")
    print("=" * 50)

    # Display audio player
    print("\nüîä Playback:")
    display(Audio(wav_path))

# Register callback
output.register_callback('notebook.process_audio', process_audio)

# All-in-one HTML + JavaScript (no external variable needed)
RECORDER_HTML = """
<div id="recorder-container" style="padding: 20px; background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); border-radius: 15px; text-align: center; color: white; font-family: Arial, sans-serif;">
    <h3 style="margin-bottom: 10px;">üé§ Bangla Speech Recording</h3>
    <p id="status" style="font-size: 18px; font-weight: bold; margin: 10px 0;">Click "Start Recording" to begin</p>
    <p id="timer" style="font-size: 32px; font-weight: bold; margin: 15px 0;">‚è±Ô∏è 0s</p>
    <div>
        <button id="startBtn"
                style="padding: 15px 30px; font-size: 18px; background: #4CAF50; color: white; border: none; border-radius: 8px; cursor: pointer; margin: 5px; box-shadow: 0 4px 6px rgba(0,0,0,0.3);">
            üé§ Start Recording
        </button>
        <button id="stopBtn" disabled
                style="padding: 15px 30px; font-size: 18px; background: #9E9E9E; color: white; border: none; border-radius: 8px; cursor: not-allowed; margin: 5px; box-shadow: 0 4px 6px rgba(0,0,0,0.3);">
            ‚èπÔ∏è Stop Recording
        </button>
    </div>
    <p id="instruction" style="font-size: 12px; margin-top: 15px; opacity: 0.8;">Allow microphone access when prompted by your browser</p>
</div>

<script>
(function() {
    const startBtn = document.getElementById('startBtn');
    const stopBtn = document.getElementById('stopBtn');
    const status = document.getElementById('status');
    const timer = document.getElementById('timer');
    const instruction = document.getElementById('instruction');

    let mediaRecorder = null;
    let chunks = [];
    let seconds = 0;
    let timerInterval = null;
    let stream = null;

    startBtn.onclick = async function() {
        try {
            // Request microphone access
            status.innerHTML = 'üé§ Requesting microphone access...';
            stream = await navigator.mediaDevices.getUserMedia({audio: true});

            // Setup recorder
            mediaRecorder = new MediaRecorder(stream, {mimeType: 'audio/webm'});
            chunks = [];

            mediaRecorder.ondataavailable = (e) => {
                if (e.data.size > 0) chunks.push(e.data);
            };

            mediaRecorder.onstop = async () => {
                // Convert to base64 and send to Python
                const blob = new Blob(chunks, {type: 'audio/webm'});
                const reader = new FileReader();
                reader.readAsDataURL(blob);
                reader.onloadend = () => {
                    const base64 = reader.result.split(',')[1];
                    google.colab.kernel.invokeFunction('notebook.process_audio', [base64], {});
                };

                // Stop all tracks
                stream.getTracks().forEach(track => track.stop());
            };

            // Start recording
            mediaRecorder.start(100); // Collect data every 100ms

            // Update UI
            status.innerHTML = 'üî¥ Recording... Speak now!';
            status.style.color = '#ff6b6b';
            startBtn.disabled = true;
            startBtn.style.background = '#9E9E9E';
            startBtn.style.cursor = 'not-allowed';
            stopBtn.disabled = false;
            stopBtn.style.background = '#f44336';
            stopBtn.style.cursor = 'pointer';
            instruction.innerHTML = 'Click "Stop Recording" when done';

            // Start timer
            seconds = 0;
            timerInterval = setInterval(() => {
                seconds++;
                timer.innerHTML = `‚è±Ô∏è ${seconds}s`;
            }, 1000);

        } catch (err) {
            status.innerHTML = '‚ùå Microphone access denied!';
            status.style.color = '#ff6b6b';
            instruction.innerHTML = 'Please allow microphone access and try again';
            console.error('Microphone error:', err);
        }
    };

    stopBtn.onclick = function() {
        if (mediaRecorder && mediaRecorder.state !== 'inactive') {
            // Stop timer
            clearInterval(timerInterval);

            // Update UI
            status.innerHTML = '‚è≥ Processing...';
            status.style.color = '#ffd93d';
            stopBtn.disabled = true;
            stopBtn.style.background = '#9E9E9E';
            stopBtn.style.cursor = 'not-allowed';
            instruction.innerHTML = 'Please wait for transcription...';

            // Stop recording
            mediaRecorder.stop();

            // Reset for next recording after delay
            setTimeout(() => {
                startBtn.disabled = false;
                startBtn.style.background = '#4CAF50';
                startBtn.style.cursor = 'pointer';
                status.innerHTML = 'Click "Start Recording" to record again';
                status.style.color = 'white';
                timer.innerHTML = '‚è±Ô∏è 0s';
                instruction.innerHTML = 'Allow microphone access when prompted by your browser';
            }, 2000);
        }
    };
})();
</script>
"""

# Display the recorder
display(HTML(RECORDER_HTML))

print("\nüëÜ Click 'Start Recording', speak in Bangla, then click 'Stop Recording'")
print("üìå Note: Allow microphone access when your browser asks for permission")
```
